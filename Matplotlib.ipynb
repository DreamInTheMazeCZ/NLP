● 라인 스타일(Line Style)
  - 문자      스타일
  - '-'   :   'solid'
  - '--'  :   'dashed'
  - '-.'  :   'dashdot'
  - ':'   :   'dotted'

```python
# 단어 입력
plt.plot(np.random.randn(50).cumsum(), linestyle = 'solid')    # Blue
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dashed')   # Orange
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dashdot')  # Green
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dotted');  # Red
```

```python
height = [np.random.randn() * i for i in range(1, 6)]
names = ['A', 'B', 'C', 'D', 'E']
y_pos = np.arange(len(names))
plt.bar(y_pos, height)
plt.xticks(y_pos, names, fontweight = 'bold')
plt.xlabel('group');
```

```python
# Stacked Bar Plot

bars1 = [12, 28, 1, 8, 22]
bars2 = [28, 7, 16, 4, 10]
bars3 = [25, 3, 23, 25, 17]
bars = np.add(bars1, bars2).tolist()

r = [0, 1, 2, 3, 4]
names = ['A', 'B', 'C', 'D', 'E']

plt.bar(r, bars1, color = 'royalblue', edgecolor = 'green')
plt.bar(r, bars2, bottom = bars1, color = 'skyblue', edgecolor = 'green')
plt.bar(r, bars3, bottom = bars2, color = 'lightblue', edgecolor = 'green')
# bottom을 활용하여 Stack 가능

plt.xlabel('group', fontweight = 'bold')
plt.xticks(r, names, fontweight = 'bold');
```

```python
# Width Stacked Bar Plot

bar_width = 0.25

bars1 = [14, 17, 9, 8, 7]
bars2 = [14, 7, 16, 4, 10]
bars3 = [21, 4, 24, 13, 17]

r1 = np.arange(len(bars1))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, bars1, width = bar_width, color = 'royalblue', edgecolor = 'green', label = 'r1')
plt.bar(r2, bars2, width = bar_width, color = 'skyblue', edgecolor = 'green', label = 'r2')
plt.bar(r3, bars3, width = bar_width, color = 'lightblue', edgecolor = 'green', label = 'r3')

plt.xlabel('group', fontweight = 'bold')
plt.xticks([r + bar_width for r in range(len(bars1))], ['A', 'B', 'C', 'D', 'E']);
plt.legend();
```

● 박스 플롯(Box Plot)

- IQR = 사분위범수 : Q1 ~ Q3 (Median = Q2 포함)
- Q1 : 중앙값과 최소값의 중간값
- Q3 : 중앙값과 최대값의 중간값
- Q1 - 1.5 X IQR
- Q3 + 1.5 X IQR
- Outlier : 사분위범수 1.5배 값에 포함되지 않는 값

```python
r1 = np.random.normal(loc = 0, scale = 0.5, size = 100)
r2 = np.random.normal(loc = 0.5, scale = 1, size = 100)
r3 = np.random.normal(loc = 1, scale = 1.5, size = 100)
r4 = np.random.normal(loc = 1.5, scale = 2, size = 100)
r5 = np.random.normal(loc = 2, scale = 2.5, size = 100)

f, ax = plt.subplots(1, 1)
ax.boxplot((r1, r2, r3, r4, r5))
ax.set_xticklabels(['r1', 'r2', 'r3', 'r4', 'r5']);
ax.set_facecolor('black')  # Subplots 배경색

# 빨간 선 : 중앙값, 범위를 벗어난 값 : Outlier
```

● X와 Y의 일관성 차트(Coherence)

```python
dt = 0.01
t = np.arange(0, 30, dt)
n1 = np.random.randn(len(t))
n2 = np.random.randn(len(t))

s1 = 1.5 * np.sin(2 * np.pi * 10 * t) + n1
s2 = np.cos(np.pi * t) + n2

plt.cohere(s1, s2 ** 2, 128, 1. / dt)
plt.xlabel('Time')
plt.ylabel('Coherence');
```

```python
plt.subplots_adjust(wspace = 1)

dt = 0.01
t = np.arange(0, 30, dt)
n1 = np.random.randn(len(t))
n2 = np.random.randn(len(t))
r = np.exp(-t/0.05)

c1 = np.convolve(n1, r, mode='same') * dt
c2 = np.convolve(n2, r, mode='same') * dt

s1 = 0.01 * np.sin(2 * np.pi * 10 * t) + c1
s2 = 0.01 * np.sin(2 * np.pi * 10 * t) + c2

plt.subplot(211)
plt.plot(t, s1, t, s2)
plt.xlim(0, 5)
plt.xlabel('Time')
plt.ylabel('s1 & s2');
plt.grid(True)

plt.subplot(212)
plt.cohere(s1, s2, 256, 1. / dt)
plt.ylabel('Coherence');

# 비교 분석
```

```python
s1 = 1.5 * np.sin(2 * np.pi * 10 * t) + c1
s2 = np.cos(np.pi * t) + c2 + np.sin(2 * np.pi * 10 * t)

fig, [ax1, ax2] = plt.subplots(2, 1)
ax1.plot(t, s1, t, s2)
ax1.set_xlim(0,5)
ax1.set_xlabel('Time')
ax1.set_ylabel('s1 & s2')
ax1.grid(True)

ax2.cohere(s1, s2, 256, 1./dt)
ax2.set_ylabel('Coherence');
```

```python
# Correlation

d1 = np.random.randn(365).cumsum()
t1 = sum(d1)
av1 = t1 / len(d1)
z1 = [i - av1 for i in d1]

d2 = np.random.randn(365).cumsum()
t2 = sum(d2)
av2 = t2 / len(d2)
z2 = [i - av2 for i in d2]

fig = plt.figure()

ax1 = fig.add_subplot(311)
ax1.plot(d1)

ax2 = fig.add_subplot(312)
ax2.plot(d2)

ax3 = fig.add_subplot(313)
ax3.xcorr(z1, z2, usevlines=True, maxlags = None, normed=True, lw=2)

plt.ylim(-1, 1);
```
